4 (30pts). Okay, enough EDA, let's get to building a model.  Using the MLP code from class (or your book), let's train the model and plot the error response.

* Start with 100 iterations and see if we can get the model to converge
* <u>Does it converge?  Does adding more iterations help?  What do you think this implies?</u>

```python
# Import needed libraries and define the perceptron class #
import numpy as np


class Perceptron:
    """Perceptron classifier.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    random_state : int
      Random number generator seed for random weight
      initialization.

    Attributes
    -----------
    w_ : 1d-array
      Weights after fitting.
    b_ : Scalar
      Bias unit after fitting.
    errors_ : list
      Number of misclassifications (updates) in each epoch.

    """
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        """ Class Constructor """
        self.eta = eta # Learning Rate #
        self.n_iter = n_iter # Number of interations through the training loop #
        self.random_state = random_state # Random initialization for the weights #

    def fit(self, X, y):
        """Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.

        Returns
        -------
        self : object

        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1]) # Initialize the weights (zero mean, 0.01 var) #
        self.b_ = np.float_(0.) # Initialize the bias unit to zero #
        
        self.errors_ = [] # Storage container for errors #

        for _ in range(self.n_iter): # Main update loop #
            errors = 0
            for xi, target in zip(X, y):
                # Implement Rosenblatt's simple update rule #
                update = self.eta * (target - self.predict(xi)) # predict function defined below #
                self.w_ += update * xi
                self.b_ += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, X): # sometimes called z (input after the summation) #
        """Calculate net input"""
        return np.dot(X, self.w_) + self.b_ # Note: this returns an accumulation for every observation in X #

    def predict(self, X): # Perform the prediction #
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, 0) # Return 1 if prediction is >= 0, return 0 otherwise #
```

```python
# Fit the model and plot the error over epochs #
percep = Perceptron(eta = 0.05, n_iter = 100) # playing around with learning rate
targets = data['Outcome']
percepData = data.drop(columns='Outcome').values

percep.fit(percepData,targets)

plt.plot(range(0, len(percep.errors_)), percep.errors_)
plt.xlabel('Epochs')
plt.ylabel('Misclassifications')
plt.title('Error over epochs')
plt.show()
```

### Solution:
* No, the model does not converge. As the iterations increase, the model does not trend consistently toward less and less misclassifications.
* Adding more iterations does not help. By turning iterations up to 500, we can observe some small trend towards less misclassifications. However, this is likely due to overtraining the model on the testing data and is not doing much in terms of improving the model for generalizations.
* This implies that the model may not be complex enough to draw the boundary it needs to for this data. An MLP should be able to handle data which is not linearly separable, but this one may not have enough layers or complexity to deal with the complexities of this data set.

5 (20pts). Let's investigate how well it does as a classifier (we won't worry too much about validation, etc. right now.)  Just get a feel for building models and training them - we'll spend some time on model evaluation in the coming weeks.

* First, this dataset didn't come with its own training vs testing set, so we'll need to create a test set of data.
* Second, we saw above that our data is imbalanced, so we can't just "randomly" grab some data to train on (we could "accidentally" only grab samples from class 0 and not train on samples from class 1)
* To deal with this, let's sort the data by class, and then do an 80/20 split on both classes: this will result in 80% of the data for training and 20% of the data for testing
    - If you set up problem 2 correct, you should already have the data separated!

* <u>How well does the classifier work?  What's the classificaiton rate on this preliminary dataset? What can we deduce from this result?</u>

```python
# Split the data into training vs. testing #

# separating in a different way from before, since before I lost the dataFrame type with the earlier method
class0 = data[data['Outcome'] == 0.0]
class1 = data[data['Outcome'] == 1.0]


# shuffling and splitting classes 80/20 using built in sample https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sample.html
class0 = class0.sample(frac=1).reset_index(drop=True) # shuffle 100% of the data and reset the indeces
zeroSplit = int(len(class0) * 0.8)
zeroTraining = class0[:zeroSplit]
zeroTesting = class0[zeroSplit:]

class1 = class1.sample(frac=1).reset_index(drop=True)
oneSplit = int(len(class1) * 0.8)
oneTraining = class1[:oneSplit]
oneTesting = class1[oneSplit:]

# put them back together and shuffle so the classes are mixed up, same strategy as before using the built-in pandas functionality
training = pd.concat([zeroTraining, oneTraining]).sample(frac=1).reset_index(drop=True)
testing = pd.concat([zeroTesting, oneTesting]).sample(frac=1).reset_index(drop=True)

# get the training and testing targets for this shuffled set
trainingY = training['Outcome']
testingY = testing['Outcome']

# get the X for the training and testing sets
trainingX = training.drop(columns='Outcome')
testingX = testing.drop(columns='Outcome')
```

```python
# Fit the model on the new training data and plot the error over epochs #
# training on 200 iterations based on email conversation
# tuning learning rate- in my testing, .1 yielded the best results
# will be redefining to make sure to start with a clean untrained perceptron
percep = Perceptron(eta = .1, n_iter = 200)
percep.fit(trainingX.values, trainingY)

plt.plot(range(0, len(percep.errors_)), percep.errors_)
plt.xlabel('Epochs')
plt.ylabel('Misclassifications')
plt.title('Error over epochs')
plt.show()
```

```python
# Predict classes #
predictions = percep.predict(testingX.values)
numberCorrect = (predictions == testingY.values).sum()
total = len(testingY)
accuracy = numberCorrect / total
print('Classification rate: ', accuracy)
```

### Solution:
* The classifier does not work amazing, which is to be expected based on what we learned in part 4. Since the model still does not converge, it is still unable to find a particularly effective "line" to draw for classification.
* It is printed above, as my randomization of the data means that the classification rate will not be the same value upon each run of the notebook.
* In my testing, the classification rate usually fell between .65 and .75.
* In my most recent test, the classification rate was about .7338, or 73.38% correct classifications.
* From these results, we can deduce that the model was able to generalize its classification capabilities to data which it was not trained on. While the model did not converge, it still got a decent amount of the predictions correct, so its boundary was at least somewhat effective. The fact that it trained without being unable to generalize means that the model was not overtrained on the training data.
