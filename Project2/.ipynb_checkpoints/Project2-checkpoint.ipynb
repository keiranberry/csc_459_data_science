1 (10pts). Let's grab the data from sklearn.datasets and do some simple exploritory data analysis (EDA).

* Note: a description of the dataset can be found [Here](https://archive.ics.uci.edu/dataset/31/covertype)
* Note: The targets are labeled [1,2,3....], go ahead and subtract 1 so the targets start at 0, i.e., [0,1,2,3...]

* <u><b>Questions:</b> How many features are there per observation? How many classes? How many observations are in the data?</u>

```python
import numpy as np
from sklearn.datasets import fetch_covtype

cov_type = fetch_covtype()

# Let's see what we're dealing with (print the shape of the data and targets) #
print("Data shape: ", cov_type.data.shape)
print("Target shape: ", cov_type.target.shape)

# Print the 10 first feature names #
print("First 10 feature names: ",cov_type.feature_names[:10])

# Grab the features and load them in a data matrix (Xt) along with their targets in the vector (y) #
Xt = cov_type.data
y = cov_type.target - 1  # subtracting one from targets as the instructions said to do, storing them in y

# Print the total number of unique classes in the target #
print("Unique classes in the target: ", len(np.unique(y))) # number of possible outcomes in the target
```

### Solution: 
* There are 54 features per observation, as we can see from the shape of the data.
* There are 7 unique classes, found by using numpy's unique function on the target vector.
* There are 581,012 observations in the data set, which can also be seen from the shape of the data.

2 (10pts). Let's get a feel for the data.  Plot a bar chart for each class (i.e., you'll get seven different bars, once for each class illustrating the total number of observations in each class)

* <u>Is the dataset balanced?  If not, what effect might this have on the model?</u>

```python
# Create and display horizontal bar chart #
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

classes, counts = np.unique(y, return_counts=True)

classLabels = [f'C{c}' for c in classes] # changing labels to match example
# adding colors to match closer to the example
cmap = ListedColormap(plt.cm.tab10.colors)

plot = plt.barh(classLabels, counts, color = cmap.colors)
plt.gca().invert_yaxis() # inverting y axis so that it is the same order as the example

plt.title('Number of observations per class')
plt.bar_label(plot)
plt.xlim(0, 350000)
plt.show()
```

### Solution:
* No, the dataset is not balanced. For example, there are less than 3000 observations of class 3, and over 283,000 observations of class 1.
* This would have an effect on the model such that it would get really good at identifying the classes with more observations, and not as good at classifying those which have less examples to train on. This could lead to incorrect categorization of smaller classes, or overtraining on the larger classes.

3 (80pts). Problem 3 is broken into different "models".  You'll receive 10pts for getting your data ready and then: Perceptron (20pts), Logistic Regression (20pts), Support Vector Machine (20pts), and Kernel SVM (10pts).  For each, you will impliment that model, and do the same analysis throughout:

* Prepare your data into a 70/30% train/test split, using stratification, and random_state = 1 - (only need to do this once)
* Scale your data to a standard distribution (zero mean and unit variance) - (only need to do this once)
* Print the bincounts for the original targets (y), training targets (y_train), and testing targets (y_test) - only need to do this once
* Fit the different models and perform some analysis:
    - Print the classification report
    - Plot the confusion matrix (it should look similar to the confusion matrix below)

```python
# Split the data into train/test with 70% train and 30% test #
from sklearn.model_selection import train_test_split

trainX, testX, trainY, testY = train_test_split(Xt, y, test_size = .3, random_state = 1, stratify = y)
```

```python
# Print the bincounts for each target array as outlined above #
print("Bincount of all targets: ", np.bincount(y))
print("Bincount of training targets: ",np.bincount(trainY))
print("Bincount of testing targets: ",np.bincount(testY))
```

```python
# Scale the data using the standard scalar option in the preprocessing module #
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
trainXScaled = scaler.fit_transform(trainX)
testXScaled = scaler.transform(testX)
```

### (20pts) Perceptron: Build the Perceptron model and perform the model evaluation (plot the confusion matrix and accuracy scores)

* <u>Discuss the results in your confusion matrix and accuracy scores.  How well does it perform?</u>

```python
# Build the Perceptron model and train it #
from sklearn.linear_model import Perceptron

perceptron = Perceptron(max_iter = 200, tol = .001)
perceptron.fit(trainXScaled, trainY)
testPredictions = perceptron.predict(testXScaled)
```

```python
# Get entire report of results, display the confusion matrix, and display the accuracy score #

# Here, I thought this meant to do all of these ^ on test data and then later predict for all samples... Talked to 
# classmates and some thought it was an overview and we were only to do those steps for testing data.
# In either case, I have done it with the testing data and have that information printed, so if anything 
# I did extra work- I am leaving the predictions on the whole set in just to be better safe than sorry.
from sklearn.metrics import * 

# get report of results
report = classification_report(testY, testPredictions)
print("Classification Report:\n", report)

# display confusion matrix
confusionMatrix = confusion_matrix(testY, testPredictions)
ConfusionMatrixDisplay(confusion_matrix = confusionMatrix).plot()
plt.title("Test Samples Confusion Matrix")
plt.show()

# display accuracy score
accuracy = accuracy_score(testY, testPredictions)
print("Accuracy on test set: ", accuracy)

# Predict for all samples #

# putting scaled samples back together to predict for all
# axis = 0 to make sure the concatenation adds more observations and not features
allXScaled = np.concatenate([trainXScaled, testXScaled], axis=0)
allY = np.concatenate([trainY, testY], axis=0)
allPredictions = perceptron.predict(allXScaled)

# Plot the confusion matrix #
confusionMatrix = confusion_matrix(allY, allPredictions)
ConfusionMatrixDisplay(confusion_matrix = confusionMatrix).plot()
plt.title("All Samples Confusion Matrix")
plt.show()
```

```python
# Display the accuracy of the model #

# I was unsure whether this should be another accuracy score or a graph,
# but based on the questions referring to "accuracy scores", I assume it 
# should be accuracy score as well.
accuracy = accuracy_score(allY, allPredictions)
print("Accuracy of model on all samples: ", accuracy)
```

### Perceptron Discussion:
* Looking at the confusion matrix, the main thing which I notice is that a large proportion of the incorrect classifications were related to classes 0 and 1, whether it be them being classified as something else or something else being classified as them. This makes sense due to the large proportion of observations in these classes. There are more of them which can be misclassified, which could be the case, but it could also be that the perceptron model has a bit of a difficult time generalizing due to the imbalanced dataset.
* Overall, the perceptron had an accuracy score of about 62%, on both the testing data and the data as a whole.
* This seems like a reasonable classification rate for perceptrons based on my findings in Project 1. Especially considering the amount of classes to differentiate between here, it clearly yields significantly better results than guessing would.

### (20 pts.) Logistic Regression: Build the Logistic regression model and perform the model evaluation (plot the confusion matrix and accuracy scores)

* Discuss which solver worked the best.  Also, add the class_weights=balanced option to see if you can "force" the model to learn the class imbalance.
* <u>Discuss the pros and cons of the different options (solvers, regulization, etc.), does it improve the accuracy or make it worse</u>
* <u>Discuss the results in your confusion matrix and accuracy scores.  How well does it perform and how does it compare to the Perceptron?</u>

```python
# Build the logistic regression model and train it (I recommend the newton-cholesky solver but feel free to play around here #
from sklearn.linear_model import LogisticRegression

logRegression = LogisticRegression(class_weight='balanced', solver='newton-cholesky', max_iter=200)
logRegression.fit(trainXScaled, trainY)
testPredictions = logRegression.predict(testXScaled)
```

```python
# Get entire report of results, display the confusion matrix, and display the accuracy score #
# get report of results
report = classification_report(testY, testPredictions)
print("Classification Report:\n", report)

# display confusion matrix
confusionMatrix = confusion_matrix(testY, testPredictions)
ConfusionMatrixDisplay(confusion_matrix = confusionMatrix).plot()
plt.title("Test Samples Confusion Matrix")
plt.show()

# display accuracy score
accuracy = accuracy_score(testY, testPredictions)
print("Accuracy on test set: ", accuracy)

# Predict for all samples #
# putting scaled samples back together to predict for all
# axis = 0 to make sure the concatenation adds more observations and not features
allXScaled = np.concatenate([trainXScaled, testXScaled], axis=0)
allY = np.concatenate([trainY, testY], axis=0)
allPredictions = logRegression.predict(allXScaled)

# Plot the confusion matrix #
confusionMatrix = confusion_matrix(allY, allPredictions)
ConfusionMatrixDisplay(confusion_matrix = confusionMatrix).plot()
plt.title("All Samples Confusion Matrix")
plt.show()
```

```python
# Display the accuracy of the model #
accuracy = accuracy_score(allY, allPredictions)
print("Accuracy of model on all samples: ", accuracy)
```

### Logistic Regression Discussion:
* In my testing, I found the Newton-Cholesky solver to work the best. Some of the others trained much longer to achieve slightly lower accuracy scores, and some did not converge for me even after spending upwards of 5 minutes waiting for the model to train.
* Different solvers work well for different data sets, such as sag and saga being faster for large datasets than liblinear is. The SciKit-Learn documentation suggested using the OneVsRestClassifier for the Newton-Cholesky solver when classifying between more than two classes, so I did a test run with that as well, but did not see any change in accuracy. Regularization helps the model with generalization ability, at the cost of some of its training accuracy. This is helpful to prevent overfitting. According to the LogisticRegression documentation, regularization is applied by default.
* The resulting accuracy score with the LogisticRegression model is slightly better than that which I achieved with the Perceptron model, by about 2%. Looking at the confusion matrix, the main difference that I observe is that the LogisticRegression model has significantly less predicted Class 0 and Class 1's which were incorrect. Where the Perceptron really struggled with that, especially given the imbalanced dataset, the Logistic Regression model seemed to do much better at dealing with that. However, the Logistic Regression model did have more troubles with misclassifying true Class 0 and Class 1 observations than the Perceptron did. This could be the result of having more of those observations to potentially misclassify, or could also be a result of the regularization, as the "noisy" data would be considered less when determining where boundaries should go. If there are lots of Class 0 and Class 1 observations that are where the model believes the boundary would be best placed for generalization, then those will be misclassified for the purpose of being able to generalize to others in the class which is being encroached upon.

```python
# Display the accuracy of the SVM model #
accuracy = accuracy_score(allY, allPredictions)
print("Accuracy of model on all samples: ", accuracy)
```

### SVM Discussion:
* Leaving the class weight empty yielded extremely similar results to having the class weight set to "balanced". Balanced class weight achieved about 37% accuracy on the test data, and leaving class weight empty achieved about 38% accuracy.
* In the confusion matrices, I can see that there is a relatively wide spread of classifications made, and not much of a trend of the diagonal having higher numbers than the other squares in the matrix. In the other two models, there was at least a trend of there being more correct guesses than incorrect, but that is not the case here. I can also see that there is a massive amount of Class 1's predicted incorrectly as Class 0. In the testing data, this makes up 60,000 of the total guesses, and in all of the data, this makes up 200,000 of the total guesses. We know there are almost 600,000 observations in the whole data set, with almost 300,000 of those being Class 1. Misclassifying most of these means being incorrect on about 1/3 of the entire data set instantly.
* These accuracy scores are much worse than the perceptron and logistic regression models. This low accuracy is still better than guessing, which would be a 1/7 chance with these 7 classes, ignoring their different weights. However, in comparison to the other models we have tested with this data, the SVM is far and away the worst one so far on this particular data set.
* Changing the class weight did not have much, if any, effect on performance. We can see this with the small 1% change that removing "balanced" caused.

### (10 pts.) Kernel SVM: Build the Kernel SVM model and perform the model evaluation (same as all the other models)

* Set kernel='rbf' and max_iter=100 so it doesn't train forever.  Also, add the class_weights=balanced option to see if you can "force" the model to learn the class imbalance.  Compare to leaving the class_weight empty.
* <u>Discuss the results in your confusion matrix and accuracy scores.  How well does it perform compared to the other models? How does changing the kernel effect the results?  What does changing gamma do?</u>

```python
# Define and fit the model #
from sklearn.svm import SVC

rbfSVM = SVC(kernel = 'rbf', max_iter = 100, gamma = .001) # removed class weights for better performance
rbfSVM.fit(trainXScaled, trainY)
testPredictions = rbfSVM.predict(testXScaled)
```

```python
# Get entire report of results, display the confusion matrix, and display the accuracy score #
# get report of results
report = classification_report(testY, testPredictions)
print("Classification Report:\n", report)

# display confusion matrix
confusionMatrix = confusion_matrix(testY, testPredictions)
ConfusionMatrixDisplay(confusion_matrix = confusionMatrix).plot()
plt.title("Test Samples Confusion Matrix")
plt.show()

# display accuracy score
accuracy = accuracy_score(testY, testPredictions)
print("Accuracy on test set: ", accuracy)

# Predict for all samples #
# putting scaled samples back together to predict for all
# axis = 0 to make sure the concatenation adds more observations and not features
allXScaled = np.concatenate([trainXScaled, testXScaled], axis=0)
allY = np.concatenate([trainY, testY], axis=0)
allPredictions = rbfSVM.predict(allXScaled)

# Plot the confusion matrix #
confusionMatrix = confusion_matrix(allY, allPredictions)
ConfusionMatrixDisplay(confusion_matrix = confusionMatrix).plot()
plt.title("All Samples Confusion Matrix")
plt.show()
```

```python
# Display the accuracy of the Kernel SVM model #
accuracy = accuracy_score(allY, allPredictions)
print("Accuracy of model on all samples: ", accuracy)
```

### Kernel SVM Discussion:
* Looking at the confusion matrices for the RBF kernel, I notice that there were much more guesses of Class 5 and 6, resulting in many more misclassifications of Classes 0 and 1. The accuracy score for this one is pretty abysmal, as 10.6% was the best that I could get with changing around gamma, C, and class weights. Removing the class weights and messing with the gamma was the only way I could get the SVM to even take a guess at all of the observations. Either way, I get the warning that it cannot converge, but this specific configuration was the only one that avoided the warnings with which the model doesn't even attempt guesses on most of the data. When it didn't attempt most of the data, it got about 9% accuracy, compared to the 10% it gets when attempting all of it. I assume the lack of class weights result in the classes with lower density getting more categorizations than they should, but it is still a better accuracy than only attempting a small amount of the data.
* Compared to the other models, this one was by far the worst with this data. This can be seen with both the accuracy scores and the confusion matrices, as it really felt like this one was just taking a stab in the dark, especially at 100 iterations. It may be better on this data with overfitting, but that would not really be helpful for anything other than boosting the accuracy score in this one instance.
* Changing the kernel did not affect the results at all how I thought it would. I assumed that the radial kernel would be much better at categorizing this complex data with many classes than the linear one was, but it did not turn out that way. After many runs with slight tweaks and a long time waiting for the model to train each time, the best I could do was 10%, which is worse than taking a 1/7 guess at one of the classes. Granted, these classes are not equal in number, so it wouldn't exactly be a 1/7 with that knowledge, but this did not do a very good job of classifying in general. As I mentioned before, I can see how the class weight may help in some cases with this, but in this case it made it worse. The ratio of correct guesses to total guesses was better, as many of the observations were not even guessed on, but when considering the correct guesses versus total guesses when not guessing is an incorrect guess, removing the class weights was ever-so-slightly better.
* Changing gamma affects how much influence a single training observation has on the decision boundary. In this case, making gamma extremely small at 0.001 was the only way I could get the RBF to take a stab at classifying all of the observations in the data set. This small of a gamma may have hindered some capabilities on individual classifications in exchange for guessing on more of the data.
